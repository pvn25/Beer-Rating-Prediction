```python


# JJ: adjective or numeral, ordinal
# JJR: adjective, comparative
# JJS: adjective, superlative
# NN: noun, common, singular or mass
# NNP: noun, proper, singular
# NNPS: noun, proper, plural
# NNS: noun, common, plural
# PRP: pronoun, personal
# PRP$: pronoun, possessive
# RB: adverb
# RBR: adverb, comparative
# RBS: adverb, superlative
# VB: verb, base form
# VBD: verb, past tense
# VBG: verb, present participle or gerund
# VBN: verb, past participle
# VBP: verb, present tense, not 3rd person singular
# VBZ: verb, present tense, 3rd person singular

markers = {"JJ" : 0,"JJR" : 0,"JJS" : 0,"NN" : 1,"NNP" : 1,"NNPS" : 1,"NNS" : 1,"PRP" : 1,"PRP$" : 1,"RB" : 2,
"RBR" : 2,"RBS" : 2,"VB" : 3,"VBD" : 3,"VBG" : 3,"VBN" : 3,"VBP" : 3, "VBZ" : 3}

dataset = []
i = 0
for l in file:
    i += 1
    line = eval(l)
    review = line['text']
    
    if i > 5000:
        break

    # Length, Verb, Adverb, Adjective, Noun    
    try:
        tokens = nltk.word_tokenize(review.lower())
        text = nltk.Text(tokens)
        tags = nltk.pos_tag(text)
        vec = [0]*6
        vec[0] = len(review)

        avgRatInBin = {}

        for t in tags:
            tag = t[1]
            if tag in markers.keys():
#                 if tag[0] == 'V':
#                     vec[1] += 1
#                 if tag[0] == 'J':
#                     vec[3] += 1
#                 if tag[0] == 'N':
#                     vec[4] += 1
                if tag[0] == 'R':
                    vec[2] += 1

        vec[5] = float(line['stars'])

    except UnicodeDecodeError:
        print "Foreign language/Error"
        
    
    k = 3 # bin size

    if(vec[2]/k not in avgRatInBin):
            avgRatInBin[vec[4]/k] = []

    avgRatInBin[vec[4]/k].append(vec[5])


    keys = avgRatInBin.keys()
    keys = sorted(keys)

    x = []
    y = []

    for key in keys:
        x.append(key)
        y.append(np.mean(avgRatInBin[key]))


    plt.scatter(np.array(x)*k, y, s = 150, color = 'navy', marker = '+')
    plt.xlabel('Number of adjectives used')
    plt.ylabel('Rating')
    print pearsonr(x, y)
    plt.show()
```
